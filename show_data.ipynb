```python
import os
import re
import numpy
import pandas as pd
# import PyPDF2
import requests
from bs4 import BeautifulSoup
from googlesearch import search
```

```python
class ResearchPaperList:
    def __init__(self): 
        self.ref_path = os.getcwd() + "/references/.txt"
        self.csv_path = os.getcwd() + "/data/Research_Paper_Data.csv"
        self.id_path = os.getcwd() + "/data/Research_Paper_List.csv"
        self.link_path = ""
        self.current_data = ""
        self.link_data = ""
        self.open_id()
        self.open_csv()
    
    def open_id(self):
        if(not os.path.exists(self.id_path)):
            print("csv file not exists")
            return
        self.id_data = pd.read_csv(self.id_path)
        return
    
    def open_csv(self):
        if(not os.path.exists(self.csv_path)):
            print("csv file not exists")
            return False
        self.csv_data = pd.read_csv(self.csv_path, index_col = 0)
        return True
    
    def open_link(self):
        if(not os.path.exists(self.link_path)):
            print("csv file not exists")
            return False
        self.link_data = pd.read_csv(self.link_path, index_col = 0)
        return True
    
    def set_data(self, paper_id):
        self.link_path = os.getcwd() + "/data/"+ str(paper_id) +".csv"
        self.open_link()
        
    def get_current_data(self):
        self.current_data = pd.DataFrame(columns=["id","title","author","date","INSPEC","DOI","citations","publisher","url","keywords"]) 
        for index  in range(len(self.link_data)):
            cur = self.link_data.iloc[index].tolist()
#             print(cur[0])
            if(cur[0] == -1):
                self.current_data.loc[index] = cur + ["","",float('nan'),"",float('nan'),"","",""]
            else:
#                 print([cur[0]]+ self.csv_data.loc[cur[0]].tolist())
#                 print(len([cur[0]]+ self.csv_data.loc[cur[0]].tolist()))
                self.current_data.loc[index] = [cur[0]] + self.csv_data.loc[cur[0]].tolist()
#         print(self.current_data)

    def get_sort_by(self,by_name,ascend):
        return self.current_data.sort_values(by=[by_name], ascending = ascend)
                
    
```

```python
rshow = ResearchPaperList()
rshow.set_data(0)
```

```python
rshow.get_current_data()
data = rshow.get_sort_by("citations", False)
data.head(300)
```
